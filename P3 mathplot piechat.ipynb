```python
import matplotlib.pyplot as plt

x = [3, 1, 3]
y = [3, 2, 1]

plt.plot(x, y)
plt.title("Line Chart")
plt.legend(["Line"])
plt.show()
```

```python
x=[3,1,3,12,2,4,4]
y=[3,2,1,4,5,6,7]
plt.bar(x,y)
plt.title("bar chat")
plt.legend(["bar"])
plt.show()
```

```python
x=[1,2,3,4,5,6,7,4]
plt.hist(x, bins=[1,2,3,4,5,6,7])
plt.title("histogram")
plt.legend(["bar"])
plt.show()
```

```python
import matplotlib.pyplot as plt

x = [3, 1, 3, 12, 2, 4, 4]
y = [3, 2, 1, 4, 5, 6, 7]

plt.scatter(x, y)
plt.title("Scatter Plot")
plt.legend(["plots"])
plt.show()

```
